#### Reading Data

*I have used pandas' read_html method to read tables from* https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M *here we have two tables and our target is to get the first one*

```python
read_data=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
read_data
```

```python
data2=read_data[1]
data2
```

```python
data=read_data[0]
print(type(data))
data
```

*So our target data is given above so I am gonna ignore the read_data[1] from the data and I'll convert the read_data[0] into DataFrame.*

```python
del(data2)
```

#### Convert the data into DataFrame

```python
data=pd.DataFrame(data)
print(type(data))
data
```

```python
data.head()
```

*Ignore the rows that have Borough as Not assigned*

```python
data=data[data['Borough']!='Not assigned']
data
```

*Assigning the Borough to the Neighbourhood, where Neighbourhood is Not assigned*

```python
data[data.loc[:,'Neighbourhood']=='Not assigned']['Neighbourhood']=data['Borough']
data
```

```python
data[data['Neighbourhood']=='Not assigned']
```

```python
data['Postal Code'].value_counts()[data['Postal Code'].value_counts()>1].sum()
```

**The code given below is used to Concat the rows in Neighbourhood that have same Postal Code**

```python
data=data.groupby(['Postal Code','Borough'])['Neighbourhood'].apply(','.join)
data=data.reset_index()
data
```
