#### To get the latitude and longitude for Toronto

```python
from geopy.geocoders import Nominatim          #Used to get the latitude and longitude of any location
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of the City of Toronto are {}, {}.'.format(latitude, longitude))
```

#### Let's Explore all the locations near to Toronto in Data

```python
import folium                   #A great visualization library for geospatial data     
#used latitude and longitude values to create Toronto map
map_tohood = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(data['Latitude'], data['Longitude'], data['Borough'], data['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=3,
        popup=label,
        color='red',
        fill=True,
        fill_color='#3199cc',
        fill_opacity=0.3,
        parse_html=False).add_to(map_tohood)  
    
map_tohood
```

#### Let's Explore the top 100 venues around The Beaches within 500 meters radius

```python
LIMIT = 100
radius = 500 
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    latitude, 
    longitude, 
    radius, 
    LIMIT)
print(url)
```

#### I am gonna use requests library to get the data from the given url in json form 

```python
import requests                      #used to get the data 
result=requests.get(url).json()
result
```

#### The given function below is used to find the name of the venue

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

#### To convert the venues detail in a tabular form

```python
from pandas.io.json import json_normalize
venues = result['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON  ..... Basically converted json into DataFrame type

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [str(col.split(".")[-1]).title() for col in nearby_venues.columns]

nearby_venues.head()
```
